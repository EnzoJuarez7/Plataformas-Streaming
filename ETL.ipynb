```python
# Agrego campo id: compondrá la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets
hulu['id'] = 'h' + hulu['show_id']
disney['id'] = 'd' + disney['show_id']
amazon['id'] = 'a' + amazon['show_id']
netflix['id'] = 'n' + netflix['show_id']
```

```python
#Agrego columna Platform
hulu['platform'] = 'hulu'
disney['platform'] = 'disney'
amazon['platform'] = 'amazon'
netflix['platform'] = 'netflix'
```

```python
#Concateno los datasets de plataformas para trabajar mejor 
plataformas_final = pd.concat([hulu,disney,amazon,netflix])
```

```python
# Concateno los datasets ratings
ratings_final = pd.concat([rating_1,rating_2,rating_3,rating_4,rating_5,rating_6,rating_7,rating_8])
```

```python
# Renombro columna 'movieId' a 'id'
ratings_final.rename(columns={'movieId': 'id'}, inplace=True)
```

```python
#Renombro la columna 'rating' por nombre correspondiente 'score' 
ratings_final.rename(columns={'rating': 'score'}, inplace=True)
```

```python
# Los valores nulos del campo rating se reemplazan por el string “G” (corresponde al maturity rating: “general for all audiences”
plataformas_final["rating"].fillna("G", inplace=True)
```

```python
# Normalización de las fechas al formato AAAA-mm-dd

# Se eliminan los espacios a los costados de la columna "date_added"
plataformas_final["date_added"] = plataformas_final["date_added"].str.strip()

# Cambia el formato de la columna "date_added" a AAAA-mm-dd
plataformas_final["date_added"] = pd.to_datetime(plataformas_final["date_added"], format='%B %d, %Y')
```

```python
# Converto la columna "timestamp" a tipo datetime y mostrar solo el día
ratings_final['timestamp'] = pd.to_datetime(ratings_final['timestamp'], unit='s').dt.strftime('%Y-%m-%d')

# Volver a convertir la columna "timestamp" a tipo datetime
ratings_final['timestamp'] = pd.to_datetime(ratings_final['timestamp'])
```

```python
#Normalización en minúsculas
plataformas_final['show_id'] = plataformas_final['show_id'].str.lower()
plataformas_final['type'] = plataformas_final['type'].str.lower()
plataformas_final['title'] = plataformas_final['title'].str.lower()
plataformas_final['director'] = plataformas_final['director'].str.lower()
plataformas_final['cast'] = plataformas_final['cast'].str.lower()
plataformas_final['country'] = plataformas_final['country'].str.lower()
plataformas_final['rating'] = plataformas_final['rating'].str.lower()
plataformas_final['duration'] = plataformas_final['duration'].str.lower()
plataformas_final['listed_in'] = plataformas_final['listed_in'].str.lower()
plataformas_final['description'] = plataformas_final['description'].str.lower()
plataformas_final['id'] = plataformas_final['id'].str.lower()
```

```python
#Normalización en minúsculas
ratings_final['id'] = ratings_final['id'].str.lower()
```

```python
# Convertimos el campo duration en dos campos: duration_int y duration_type. 
# El primero en un integer y el segundo en un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

#Se eliminan los espacios laterales en la columna "duration"
plataformas_final['duration'] = plataformas_final['duration'].str.rstrip()
plataformas_final['duration'] = plataformas_final['duration'].str.lstrip()
```

```python
#Separo la cantidad de minutos con la palabra min
separado_data_final = plataformas_final['duration'].str.split(' ', n=1, expand=True)
plataformas_final['duration_int'] = separado_data_final[0]
plataformas_final['duration_type'] = separado_data_final[1]
```

```python
#Normalización de seasons a season
plataformas_final['duration_type'] = plataformas_final['duration_type'].str.replace("seasons", "season", case=False)
```

```python
#Reemplaza los valores NaN por 0 para poder modificar el Dtype de duration_int a intg64
plataformas_final['duration_int'].fillna(0, inplace=True)
plataformas_final['duration_int'] = plataformas_final['duration_int'].astype('int64')
```

```python
data_final = pd.merge(plataformas_final, ratings_final, on='id')
```

```python
#Elimino la columna 'duration'
data_final.drop(columns='duration', inplace=True)
```

```python
#Elimino la columna 'show_id'
data_final.drop(columns='show_id', inplace=True)
```
